5. Taking Iris data, plot the following with proper legend and axis labels: (Download IRIS data from:https://archive.ics.uci.edu/ml/datasets/iris or import it from sklearn.datasets)


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load Iris dataset
iris = load_iris()
iris_data = iris.data
iris_target = iris.target
iris_feature_names = iris.feature_names
iris_target_names = iris.target_names
```

```python
iris_df = pd.DataFrame(data=iris_data, columns=iris_feature_names)
iris_df['Species'] = iris_target_names[iris_target]
```

a. Plot bar chart to show the frequency of each class label in the data.

```python
plt.figure(figsize=(8, 5))
sns.countplot(x='Species', data=iris_df)
plt.title('Frequency of Each Class Label in Iris Dataset')
plt.xlabel('Species')
plt.ylabel('Frequency')
plt.show()
```

b. Draw a scatter plot for Petal width vs sepal width.

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(x='petal width (cm)', y='sepal width (cm)', hue='Species', data=iris_df)
plt.title('Scatter Plot: Petal Width vs Sepal Width')
plt.xlabel('Petal Width (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend()
plt.show()
```

c. Plot density distribution for feature petal length.


```python
plt.figure(figsize=(8, 5))
sns.kdeplot(data=iris_df, x='petal length (cm)', hue='Species', fill=True)
plt.title('Density Distribution: Petal Length')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Density')
plt.legend()
plt.show()
```

d. Use a pair plot to show pairwise bivariate distribution in the Iris Dataset.

```python





plt.figure(figsize=(12, 8))
sns.pairplot(iris_df, hue='Species', markers=["o", "s", "D"])
plt.suptitle('Pairwise Bivariate Distribution in Iris Dataset', y=1.02)
plt.show()
```
