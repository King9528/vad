3. Create a dataframe having at least 3 columns and 50 rows to store numeric data generated using a random function. Replace 10% of the values by null values whose index positions are generated using random function. Do the following:


```python
import numpy as np 
import pandas as pd
```

```python
# # Set a seed for reproducibility
# np.random.seed(42)

# Create a dataframe with 3 columns and 50 rows of random numeric data
data = {'Column1': np.random.randn(50),
        'Column2': np.random.randn(50),
        'Column3': np.random.randn(50)}

df = pd.DataFrame(data)

# Replace 10% of the values with null values
null_indices = np.random.choice(df.size, size=int(0.1 * df.size), replace=False)
df.values.flat[null_indices] = np.nan
df
```

a. Identify and count missing values in a dataframe.


```python
missing_values = df.isnull().sum()
print("Missing values:\n", missing_values)
```

b. Drop the column having more than 5 null values.


```python
df = df.dropna(axis=1, thresh=len(df) - 5)
df
```

c. Identify the row label having maximum of the sum of all values in a row and drop that row.


```python
max_sum_row = df.sum(axis=1).idxmax()
df = df.drop(max_sum_row)
df
```

d. Sort the dataframe on the basis of the first column.


```python
# first_column_name = df.columns[0]
df = df.sort_values("Column1")
df
```

e. Remove all duplicates from the first column.


```python
# first_column_name = df.columns[0]
df = df.drop_duplicates("Column1")
df
```

column.


```python
# correlation = df['Column1'].corr(df['Column2'])
# covariance = df['Column2'].cov(df['Column3'])
# print("Correlation between Column1 and Column2:", correlation)
# print("Covariance between Column2 and Column3:", covariance)

#---------------------------------------------------------------------------------------------------

# first_column_name = df.columns[0]
# second_column_name = df.columns[1]
# third_column_name = df.columns[2]

# correlation_col1_col2 = df[first_column_name].corr(df[second_column_name])
# covariance_col2_col3 = df[second_column_name].cov(df[third_column_name])

# print(f"f. Correlation between {first_column_name} and {second_column_name}: {correlation_col1_col2}")
# print(f"   Covariance between {second_column_name} and {third_column_name}: {covariance_col2_col3}")

#---------------------------------------------------------------------------------------------------

if df.shape[1] >= 3:
    first_column_name = df.columns[0]
    second_column_name = df.columns[1]
    third_column_name = df.columns[2]

    correlation_col1_col2 = df[first_column_name].corr(df[second_column_name])
    covariance_col2_col3 = df[second_column_name].cov(df[third_column_name])

    print(f"f. Correlation between {first_column_name} and {second_column_name}: {correlation_col1_col2}")
    print(f"   Covariance between {second_column_name} and {third_column_name}: {covariance_col2_col3}")
else:
    print("Not enough columns for correlation and covariance calculations.")
    
```

g. Detect the outliers and remove the rows having outliers.


```python
z_scores = (df - df.mean()) / df.std()
outliers = (z_scores > 3) | (z_scores < -3)
df = df[~outliers.any(axis=1)]
df
```

h. Discretize second column and create 5 bins

```python
df['Column2_bins'] = pd.cut(df['Column2'], bins=5)
df
```
